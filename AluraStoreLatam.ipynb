LAS COORDENADAS SON IDENTICAS PARA CADA TIENDA, NO HAY DIFERENICA EN SU COMPOSICION DE LATITUD Y LONGITUD

```python
u1 = tienda1[['lat','lon']].drop_duplicates()
u2 = tienda2[['lat','lon']].drop_duplicates()
u3 = tienda3[['lat','lon']].drop_duplicates()
u4 = tienda4[['lat','lon']].drop_duplicates()

print("Coordenadas únicas T1:", len(u1))
print("Coordenadas únicas T2:", len(u2))
print("Coordenadas únicas T3:", len(u3))
print("Coordenadas únicas T4:", len(u4))

# Coincidencias
c12 = pd.merge(u1, u2, on=['lat','lon']).shape[0]
c13 = pd.merge(u1, u3, on=['lat','lon']).shape[0]
c14 = pd.merge(u1, u4, on=['lat','lon']).shape[0]

print("\nCoincidencias reales T1 vs T2:", c12)
print("Coincidencias reales T1 vs T3:", c13)
print("Coincidencias reales T1 vs T4:", c14)

# Porcentajes reales
p12 = (c12 / len(u1)) * 100
p13 = (c13 / len(u1)) * 100
p14 = (c14 / len(u1)) * 100

print(f"\n% coincidencia T1 vs T2: {p12:.2f}%")
print(f"% coincidencia T1 vs T3: {p13:.2f}%")
print(f"% coincidencia T1 vs T4: {p14:.2f}%")
```

## Rendimiento

Haciendo uso de clustering

```python
from sklearn.cluster import KMeans
import pandas as pd

t1 = tienda1[['lat','lon']].assign(tienda='T1')
t2 = tienda2[['lat','lon']].assign(tienda='T2')
t3 = tienda3[['lat','lon']].assign(tienda='T3')
t4 = tienda4[['lat','lon']].assign(tienda='T4')

data = pd.concat([t1,t2,t3,t4], ignore_index=True)

kmeans = KMeans(n_clusters=6, random_state=42)
data['region'] = kmeans.fit_predict(data[['lat','lon']])

data.head()
```

REGIONES SELLECIONADAS

```python
region_counts = data.groupby(['region','tienda']).size().reset_index(name='ventas')

region_pivot = region_counts.pivot(index='region', columns='tienda', values='ventas').fillna(0)
region_pivot
```

GRAFICO DE REGIONES

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
scatter = plt.scatter(data['lon'], data['lat'], c=data['region'], cmap='tab10', alpha=0.6)

plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Regiones Geográficas Detectadas por K-Means")
plt.colorbar(scatter, label='Región')
plt.show()
```

RENDIMIENTO POR REGIÓN

```python
region_pivot.plot(kind='bar', figsize=(10,6))
plt.title("Rendimiento por Región (Ventas por Tienda)")
plt.ylabel("Número de Ventas")
plt.xlabel("Región")
plt.xticks(rotation=0)
plt.show()
```
